```python
#creditApprovalUCI.csv를 credit3라는 이름으로 데이터프레임 읽으세요
#컬럼별 결측 확인해보세요
import pandas as pd
credit3 = pd.read_csv("https://raw.githubusercontent.com/ws-l/hh_2026_1/main/data/creditApprovalUCI.csv")
credit3.describe() #요약  
#credit.default10yr.value_counts() 
credit3.isnull().mean()
credit3.fillna(0)   
credit3.dropna()
credit3.select_dtypes(exclude="object").fillna(0) 
credit3.select_dtypes(include="object")
```

```python
numeric = credit3.select_dtypes(exclude="object")
categorical = credit3.select_dtypes(include="object")

from sklearn.impute import SimpleImputer
imp_n = SimpleImputer(strategy="mean")
imputed_n = imp_n.fit_transform( numeric )   
imp_c = SimpleImputer(strategy="most_frequent")
imputed_c = imp_c.fit_transform( categorical )   

numeric_imputed = pd.DataFrame(imputed_n, columns = numeric.columns)
categorical_imputed = pd.DataFrame(imputed_c, columns = categorical.columns)

pd.concat( [numeric_imputed, categorical_imputed], axis=1 )  
```

```python
import numpy as np
import pandas as pd
def get_first_cabin(row):
	try:
		return row.split()[0]	#공백을 기준으로 나눈 첫번째 값
	except:
		return np.nan
data = pd.read_csv("https://raw.githubusercontent.com/ws-l/hh_2026_1/main/data/titanic_raw.csv")
data = data.replace('?', np.nan)	#치환
data.replace(["female", "male"], ["f", "m"])     
data.replace({"female":"f", "male":"m"}, inplace=True)
data["home.dest"].replace("Montreal", "몽트레알", regex=True)   
data.cabin 
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/ws-l/hh_2026_1/main/data/crx.data', header=None)
varnames = ['A'+str(s) for s in range(1,17)]
data.columns = varnames
data.head()
# A16 컬럼-> value_counts()
# + Positive, - Negative    ->replace
# tive->@@@@      ->replace, regex=True
# A2변수는 각 값을 3을 곱하고 10으로 나눈 값을 변경  ->함수, apply
#data["A16"].value_counts()
data["A16"].replace({"+": "Positive", "-":"Negative"}, 
                    inplace=True  )
data["A16"].replace( "[A-Z]", "@", regex=True)  
def func1(x):
    try:
        return (float(x)*3)/10
    except:
        return np.nan
data.select_dtypes(exclude="object").applymap(func1)  
```

```python
#!pip install seaborn
import seaborn as sns
data = sns.load_dataset("flights")
data.groupby("month")["passengers"].mean()
data2 = data.pivot(columns="month", index="year",
            values="passengers")
data.pivot_table(columns="month", index="year",
            values="passengers", aggfunc="mean")  
sns.heatmap(data2)      

#creditset2.csv를 읽으셔서 pivot(또는 pivot_table)
#age 기준으로 30대 이하, 40대 이상
#default여부와 30대/40대 구분으로 pivot_table
#대상 value는 loan/income비율 계산하셔서 평균
credit = pd.read_csv('https://raw.githubusercontent.com/ws-l/hh_2026_1/main/data/creditset2.csv')
credit.default10yr.replace([0,1], ["normal","default"],
                            inplace=True)
def func2(x):
    if x < 40:
        return "30s"
    else:
        return "40s"
credit["age2"] = credit["age"].apply( func2 )
credit["LTI"] = credit.loan / credit.income
credit.pivot_table(columns="age2", index="default10yr",
            values="LTI", aggfunc="mean")  
```
