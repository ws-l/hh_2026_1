```python
#함수, 데이터프레임 apply / applymap
#crx.data를 읽어주세요
data = pd.read_csv("https://raw.githubusercontent.com/ws-l/hh_2026_1/main/data/crx.data",
                   header=None)
data.columns = ['A' + str(i)    for i in range(1,17)]
#2,3,8,11,14,15
data["A2"] = data["A2"].replace('?', np.nan).astype("float")
#float 컬럼들만 선택
data2 = data.select_dtypes(include="float")
#10보다 큰 경우 3을 곱하고, 10보다 작은 경우에는 5더하기
#A3에 적용해주세요.
def func1(x):
    if x > 10:
        return (x*3)
    else:
        return (x+5)
data2.A3.apply( func1 )   
```

```python
import pandas as pd
data = pd.read_csv("c:\\Users\\owner\\Downloads\\data.csv")
```

```python
data.groupby("Pass.Fail")["X20"].mean() 

#X86의 값이 2.4보다 크면 large, 아니면 small이 되도록 
#apply에 함수를 이용해서 값을 변환해서 new1이라는 컬럼...
def func1(x):
    if x>2.4:
        return 'large'
    else:
        return 'small'
data["new1"]= data.X86.apply(func1)
#Pass.Fail과 new1을 그룹바이에 넣으셔서 X87에 대한 합계....
data.groupby(["Pass.Fail", "new1"])["X87"].sum()
data.pivot_table(index="Pass.Fail", columns="new1",
            values="X87", aggfunc=["mean", "sum"] )  
```

```python
credit = pd.read_csv("creditset2.csv" )
credit["LTI"] = credit.loan/credit.income
credit.LTI.mean()#0.098
def func3(x):
    if x>0.098 :
        return 1
    else:
        return 0
credit["new2"] = credit.LTI.apply(func3)
credit.groupby(["default10yr", "new2"])["age"].mean()
credit.pivot_table(index="default10yr", columns="new2",
                   values="age", aggfunc=["mean","min"])
```
