```python
#creditApprovalUCI.csv을 읽으셔서 결측을 체크해보세요
credit = pd.read_csv("https://raw.githubusercontent.com/ws-l/hh_2026_1/main/data/creditApprovalUCI.csv")  
credit.isnull().mean().sort_values(ascending=False) #결측치 정렬
credit.dropna() #결측있는 행 삭제, inplace=True
#imputing
c_n = credit.select_dtypes(exclude="object")#.fillna(0)
c_o = credit.select_dtypes(include="object")#.fillna("NN")  

from sklearn.impute import SimpleImputer
imputer_n =  SimpleImputer(strategy="median") #mean, median
imputed_n = imputer_n.fit_transform( c_n ) 
d1 = pd.DataFrame( imputed_n, columns=c_n.columns )

imputer_o =  SimpleImputer(strategy="most_frequent") 
imputed_o = imputer_o.fit_transform( c_o ) 
d2 = pd.DataFrame( imputed_o, columns=c_o.columns )
credit2 = pd.concat(  [d1, d2], axis=1) 
```
